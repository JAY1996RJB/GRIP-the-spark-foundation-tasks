# Exploratory Data Analysis - Sports

Objective : find out the most successful teams, players and factors 
            contributing win or loss of a team.
            
# Pandas Profiling

```python
import pandas as pd
import numpy as np
from pandas_profiling import ProfileReport
```

```python
deliveries = pd.read_csv('deliveries.csv')
```

```python
matches = pd.read_csv('matches.csv')
```

```python
deliveries.head(2)
```

```python
matches.head(2)
```

```python
data = pd.merge(deliveries,matches,left_on='match_id',right_on='id')
```

```python
data.info()
```

```python
data.describe()
```

```python
profile = ProfileReport(data,title='Exploratory Data Analysis - Sports',explorative=True)
```

```python
profile.to_file("output of sports.html")
```
