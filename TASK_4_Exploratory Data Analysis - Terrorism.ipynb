```python
data= data[['gname','weaptype1_txt','targtype1', 'targtype1_txt','iyear', 'imonth', 'iday','country', 'country_txt', 'region', 'region_txt','provstate', 'city','attacktype1','attacktype1_txt','nkill','nwound','weaptype1']]
data.head()
```

```python
profile = ProfileReport(data,title='Exploratory Data Analysis - Terrorism',explorative=True)
```

```python
profile.to_file("output of Terrorism.html")
```

```python
hotspot_report = sv.analyze([data,"Terrorism"],target_feat='iyear')
```

```python
hotspot_report.show_html('hotspot_report.html',layout='vertical',scale=1.0)
```
